Dans notre cas, nous souhaitons retrouver les distributions de mots sur les différents thèmes, les différentes proportions de thèmes pour chaque document, les proportions d’apparition d’un thème sur le corpus. Tout cela à partir des différents documents. Ce qui nous permet par la suite de déterminer le thème d’un document, les mots les plus associés à certains thèmes, etc
    
**Modélisation**  
    
1. Extraction de features (à l'aide de topic models non-paramétriques) : 
    * bag of words
    * TF-IDF &rarr; **Oui, plus intuitif**  
    
1. Word Embedding :
    * [LSA (Latent Semantic Analysis) / LSI (Latent Semantic Indexing)](https://docs.scipy.org/doc/scipy/reference/generated/scipy.sparse.linalg.svds.html) : Basé sur la SVD (Singular Value Decomposition)
        * A (term-document) décomposée (**A = USV'**) en 
        * U (term-concept), S (singular value), V (concept-document)  
    Difficile à interpréter &rarr; **NFM ou pLSA**
    
    * [NMF (Non-negative Matrix Factorization)](https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.NMF.html) : &rarr; **Oui** 
    
    * pLSA : problème de surapprentissage &rarr; **Non, solution LDA**
    
    * [LDA (Latent Dirichlet Allocation)](http://devdoc.net/python/sklearn-0.18/modules/generated/sklearn.decomposition.LatentDirichletAllocation.html) : Topic Distribution in a corpus  
    Problème : Certains thèmes sont plus proches que d’autres. L’hypothèse d’utiliser la distribution de Dirichlet considère que les différents thèmes sont complètement indépendants alors qu’en réalité certains thèmes ont en général plus de chances d’apparaître ensemble. &rarr; **Oui**  
    
    * ASUM (Aspect Sentiment Unification Model) : Dérivé de LDA
    * VODUM (Viewpoint and Opinion Discovery Unification Model) : Dérivé de LDA
    * TOT (Topics over Time) : Dérivé de LDA
    * GLDA (Gaussian Latent Dirichlet Allocation) : Dérivé de LDA
    
    * CBOW pour déterminer le contexte (tags) en fonction des mots
    * Neural Networks  &rarr; **Non, pas de NN dans la première phase** :
        * word2vec ([word2vec perso](https://radimrehurek.com/gensim/auto_examples/))
        * [Lda2Vec](https://github.com/cemoody/lda2vec) : LDA + word2vec
    * [GloVe](https://nlp.stanford.edu/projects/glove/) ? Swivel ?
    * [FastText](https://fasttext.cc/) ?  
    
1. Modélisation :
    * Modèle personnel ($\text{Words}\times\text{Tags}\rightarrow A$ ; $\text{Tags} = \text{Words}\times A$) : **Oui, Simple/Base**
    * [Naive Bayes](http://www.nltk.org/_modules/nltk/classify/naivebayes.html) ?
        * faible variance, généralisation rapide
        * plus grand biais
        * plus efficace pour #individus faible et #features faibles
    * Regression logistique : Non car classification discriminante ?
    * [SVD](https://multithreaded.stitchfix.com/blog/2017/10/18/stop-using-word2vec/) ? +RBF kernel ?
    * NN
        * LSTM : [explication](https://colah.github.io/posts/2015-08-Understanding-LSTMs/), [implementation](https://github.com/karpathy/char-rnn)

```python
def display_topics(model, feature_names, no_top_words=10):
    for topic_idx, topic in enumerate(model.components_):
        print("Topic {}:".format(topic_idx))
        print(" ".join([feature_names[i] for i in topic.argsort()[:-no_top_words - 1:-1]]))

def cumulate_scores(mdf, name):
    global rdf
    try:
        rdf = rdf.append(pd.concat([mdf], keys=[name]))
    except:
        rdf = pd.concat([mdf], keys=[name])

def df_scores(NT=None):
    import warnings
    warnings.filterwarnings("ignore")
    scores = {
        'Train': {
            'Accuracy' :  skl.metrics.accuracy_score(Y_train[:NT], Y_pred_train),
            'Recall' :    skl.metrics.recall_score(Y_train[:NT], Y_pred_train, average='weighted'),
            'Precision' : skl.metrics.precision_score(Y_train[:NT], Y_pred_train, average='weighted'),
            'F1 score' :  skl.metrics.f1_score(Y_train[:NT], Y_pred_train, average='weighted'),
            'Jaccard' :   skl.metrics.jaccard_score(Y_train[:NT], Y_pred_train, average='weighted')},
        'Test': {
            'Accuracy' :  skl.metrics.accuracy_score(Y_test[:NT], Y_pred_test),
            'Recall' :    skl.metrics.recall_score(Y_test[:NT], Y_pred_test, average='weighted'),
            'Precision' : skl.metrics.precision_score(Y_test[:NT], Y_pred_test, average='weighted'),
            'F1 score' :  skl.metrics.f1_score(Y_test[:NT], Y_pred_test, average='weighted'),
            'Jaccard' :   skl.metrics.jaccard_score(Y_test[:NT], Y_pred_test, average='weighted')}}
    return pd.DataFrame.from_dict(scores).T
```

```python
class SimpleModel(skl.base.RegressorMixin):
    def __init__(self, threshold=None):
        self.threshold = threshold
        
    def fit(self, X, Y):
        if isinstance(X, np.ndarray) and isinstance(Y, np.ndarray):
            self.A_ = X.T.dot(Y)
        else:
            self.A_ = X.T * Y
        
    def predict(self, X):
        if isinstance(X, np.ndarray) and isinstance(self.A_, np.ndarray):
            continuous = X.dot(self.A_)
        else:
            continuous = X * self.A_
           
        if self.threshold is None:
            return continuous
        else:
            normalized = (continuous.T / continuous.max(1).toarray().T).T
            return normalized > self.threshold
        
    # def predict(self, X):
    #     if isinstance(X, np.ndarray) and isinstance(self.A_, np.ndarray):
    #         continuous = X.dot(self.A_) 
    #         normalized = (continuous.T / continuous.max(1)).T
    #         return normalized > self.threshold
    #     else:
    #         from scipy import sparse
    #         continuous = sparse.csr_matrix((X * self.A_))
    #         normalized = (continuous.T / continuous.max(1).todense().T).T
    #         return sparse.csr_matrix(normalized > self.threshold)
```

## Binary Relevance
C'est la technique la plus simple, qui traite basiquement chaque étiquette par un problème de classification de classe unique.
Prenons une variable indépendante $x$ et $y1,y2,y3$ comme les étiquettes de la variable dépendante.
Donc, ce que la **pertinence binaire** fait, c'est de traiter chaque variable indépendante comme une classe distincte en considération de la variable indépendante.  
Il mappe donc $x →y1$ et $x →y2$ et $x →y3$

### Gaussian Naive Bayes

```python
%%time
# Temps de calcul 1/5. Estimation sur 35000 : 2H
NN = 1000
model = problem_transform.BinaryRelevance(skl.naive_bayes.GaussianNB())
model.fit(X_train[:NN], Y_train[:NN])
```

## Classifier Chain
Ici, le premier classificateur est entraîné uniquement sur les données d'entrée, puis chaque classificateur suivant est entraîné sur l'espace d'entrée et tous les classificateurs précédents de la chaîne.
Disons que nous avons $x$ comme variable indépendante et $y1,y2,y3$ comme étiquettes de variable dépendante.
Nous avons donc essentiellement 3 sous-ensembles de classification $x →y1$ et $x →y1,y2$ et $x →y1,y2,y3$.
